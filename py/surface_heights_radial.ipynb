```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
def get_path(a1, a2):
    return Path(f"./surface_heights_radial_{a1}_random.txt")
```

```python
paths = [get_path(a, 0) for a in range (40, 41)]
paths
```

```python
vdata = [np.loadtxt(p, skiprows=2) for p in paths]
```

```python
def plot(i):
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.plot(np.deg2rad(vdata[i][:,0]), vdata[i][:,1], label=f"{40} deg")
    ax.legend(loc='upper right')
    ax.set_title("0 K 8 keV 0 azimuth")
```

```python
plot(0)
```

```python
plot(1)
```

```python
plot(2)
```

```python
plot(3)
```

```python
new_40 = Path(f"./surface_heights_radial_{40}_random.txt")
old_40 = Path(f"./surface_heights_radial_{40}.txt")
sw_40 = Path(f"./surface_heights_radial_{40}_sw.txt")
new_data = np.loadtxt(new_40, skiprows=2)
old_data = np.loadtxt(old_40, skiprows=2)
sw_data = np.loadtxt(sw_40, skiprows=2)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(np.deg2rad(old_data[:,0]), old_data[:,1], label=f"{40} deg old")
ax.plot(np.deg2rad(new_data[:,0]), new_data[:,1], label=f"{40} deg new")
ax.plot(np.deg2rad(sw_data[:,0]), sw_data[:,1], label=f"{40} deg sw")
ax.legend(loc='upper right')
ax.set_title("0 K 8 keV 0 azimuth")
```
